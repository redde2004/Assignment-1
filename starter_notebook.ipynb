````python
#STEP 2

print("Describe")
print(df.describe())
#TO get a numeric Summary

print()

print("Attrition!")
print(df['Attrition'].value_counts())
#create a visualization to see how imbalanced the dataset is

print()

print("get Dummies")
df_encoded = pd.get_dummies(df, drop_first=True)
#encode categorical variables into numeric format

print()

print("X cleaned attrition")
X_cleaned = df_encoded.drop('Attrition_Yes', axis=1)
y = df_encoded['Attrition_Yes']
#Create A feature matrix
````

````python
#Step 3

#split the data into training and test sets with an 80/20 split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_cleaned, y, test_size=0.2, stratify=y, random_state=42)

#Skill Trees with random State of 42
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

#Predition
y_pred_dt = dt_model.predict(X_test)

#Metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score

dt_accuracy = accuracy_score(y_test, y_pred_dt)
dt_precision = precision_score(y_test, y_pred_dt)
dt_recall = recall_score(y_test, y_pred_dt)

#Print Results
print("Baseline Decision Tree Results")
print("------------------------------")
print(f"Accuracy: {dt_accuracy:.3f}")
print(f"Precision: {dt_precision:.3f}")
print(f"Recall: {dt_recall:.3f}")
````

````python
#Step 4

#Train Random Forest
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=200, max_depth=None, min_samples_split=10, min_samples_leaf=2, max_features='sqrt', class_weight='balanced', random_state=42)
rf_model.fit(X_train, y_train)

#Calculate Probabilities Predictions
y_proba_rf = rf_model.predict_proba(X_test)[:, 1]

#Threshold Added In
y_pred_rf = (y_proba_rf >= 0.35).astype(int)

#Metrics
rf_accuracy = accuracy_score(y_test, y_pred_rf)
rf_precision = precision_score(y_test, y_pred_rf)
rf_recall = recall_score(y_test, y_pred_rf)

#Print!
print("Random Forest Results")
print("---------------------")
print(f"Accuracy: {rf_accuracy:.3f}")
print(f"Precision: {rf_precision:.3f}")
print(f"Recall: {rf_recall:.3f}")

#Model Comparison
import pandas as pd

model_comparison = pd.DataFrame({
    'Model': ['Decision Tree', 'Random Forest'],
    'Accuracy': [dt_accuracy, rf_accuracy],
    'Precision': [dt_precision, rf_precision],
    'Recall': [dt_recall, rf_recall]})

print()
print("MODEL COMPARISON!")
model_comparison
````
